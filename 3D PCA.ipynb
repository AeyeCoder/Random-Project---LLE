```python
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x,t=make_swiss_roll(n_samples=50000)
fig=plt.figure(figsize=(12,4))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],c=t,cmap='magma')
plt.title("Swiss Roll (Original)")
plt.show()
```

```python
from sklearn.random_projection import GaussianRandomProjection
from sklearn.manifold import locally_linear_embedding

rp=GaussianRandomProjection(n_components=2)
x_rp=rp.fit_transform(x)
x_lle,_=locally_linear_embedding(x,n_components=2,n_neighbors=20)
def plot_2d(name,data):
    plt.figure(figsize=(5,3))
    plt.scatter(data[:,0],data[:,1],c=t,cmap='magma')
    plt.title(f"Unrolled through {name}")
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.grid(True)
    plt.show()

plot_2d("Random Projection",x_rp)
plot_2d("Local Linear Embedding",x_lle)
```
